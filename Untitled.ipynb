Generate data of people's mass in pounds

```python
np.random.seed(42)
normal_male_mass = np.random.normal(loc=181, scale=24, size=6000)
normal_female_mass = np.random.normal(loc=132, scale=22, size=6500)

all_mass_values = np.concatenate(
    (
        normal_male_mass,
        normal_female_mass
    ),
    axis=0,
)

df_ppl_mass = pd.DataFrame(data={'ppl_mass_pounds': all_mass_values})
```

```python
df_ppl_mass.head()
```

view distribution people mass

```python
sns.histplot(
    df_ppl_mass['ppl_mass_pounds'],
    kde=True,
    color='skyblue',
)

plt.xlabel("mass [pounds]", labelpad=14)
plt.ylabel("probability of occurence", labelpad=14)
plt.title("Distribution of Normal Mass People", y=1.015, fontsize=20)
```

```python
#Calculation population mean
pop_mean_mass = df_ppl_mass['ppl_mass_pounds'].mean()
pop_mean_mass
```

```python
pop_std_mass = df_ppl_mass['ppl_mass_pounds'].std()
pop_std_mass
```

```python
sample_means = []
n = 25
for sample in range (0, 300):
    sample_values = np.random.choice(
        a=df_ppl_mass['ppl_mass_pounds'],
        size=n
    )
    sample_mean = np.mean(sample_values)
    sample_means.append(sample_mean)
```

```python
sns.histplot(
    sample_means,
    kde=True,
    color='skyblue'
)
plt.title("Distrbution of Sample Means ($n=25$) of People's Mass in Pounds", y=1.015, fontsize=20)
plt.xlabel("sample mean mass [pounds]", labelpad=14)
plt.ylabel("frequency of occurence", labelpad=14)
```

Calculate Sampling Distribution Summary Statistics

```python
median_of_sample = np.median(sample_means)
median_of_sample
```

```python
mean_of_sample = np.mean(sample_means)
mean_of_sample
```

```python
pop_mean_mass
```

```python
std_of_sample = np.std(sample_means)
std_of_sample
```

Equation of Standard Deviation of Sampling Distribution

```python
standard_error = pop_std_mass/np.sqrt(n)
standard_error
```

```python
std_of_sample
```

Critical Z Score

```python
tail = 1 - norm.cdf(1.96)
tail
```

```python
1 - (tail * 2)
```

```python
lower_95 = mean_of_sample - 1.96 * standard_error
lower_95
```

```python
upper_95 = mean_of_sample + 1.96 * standard_error
upper_95
```

```python
from scipy import stats

kde = stats.gaussian_kde(sample_means)
pos = np.linspace(np.min(sample_means), np.max(sample_means), 10000)
plt.plot(
    pos,
    kde(pos),
    color='teal'
)

shade = np.linspace(lower_95, upper_95, 300)
plt.fill_between(
    shade,
    kde(shade),
    alpha=0.45,
    color='teal'
)

plt.text(
    x=154,
    y=0.01,
    horizontalalignment='center',
    fontsize=17,
    s="95% of sample menas fall within\nthis shaded area of plus or minus\n1.96 z-scores from the mean  ",
    bbox=dict(
        facecolor='whitesmoke',
        boxstyle="round, pad=0.1",
    )
)

plt.title("Distribution of Sample Means ($n=25$) of People's Mass in Pounds", y=1.015, fontsize=20)
plt.xlabel("sample mean in mass [pounds]", labelpad=14)
plt.ylabel("frequency of occurence", labelpad=14)
```

# Hypothesis Testing for One Proportion

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
p_us = len(df[df['target'] == 1]) / len(df)
p_us
```

```python
se = np.sqrt(0.42  * (1 - 0.42) / len(df))
se
```

```python
be = p_us
he = 0.42
test_stat = (be - he)/se
test_stat
```

```python
pvalue = 2 * dist.norm.cdf(-np.abs(test_stat))
pvalue
```
